```python
def train_ridge( x_train: np.ndarray, y_train:np.ndarray, x_val: np.ndarray, y_val: np.ndarray, max_iter: int = int(1e8))-> dict:
    '''
        Instantiate an object of Ridge Regression, train the model object using training data for the given N-bootstraps iterations.
        In each iteration train the model for all alpha_vals, compute the auc values with the validation set and append the values to aucs_ridge.
        Return the "ridge_mean_auc" dictionary which stores the mean auc for each alpha.
    '''
    n_bootstraps = int(1e3)
    alpha_vals = [1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3]
    aucs_ridge = {alpha:[] for alpha in alpha_vals}
    for _ in range(n_bootstraps):
        #resampling the training data set
        x_train_sample, y_train_sample = resample(x_train, y_train)
        for alpha in alpha_vals:
            ridge = Ridge(alpha=alpha, max_iter=max_iter)
            ridge.fit(x_train_sample, y_train_sample)
            y_pred = ridge.predict(x_val)

            ridge_aucs = roc_auc_score(y_val, y_pred)
            aucs_ridge[alpha].append(ridge_aucs)  

    ridge_mean_auc = {alpha: np.mean(aucs) for alpha, aucs in aucs_ridge.items()}
    
    return ridge_mean_auc

#train the lasso model and fetch the dictionary containing the mapping of alpha and aucs
ridge_mean_auc = train_ridge(np.array(X_train), np.array(y_train), np.array(X_valid), np.array(y_valid))

optimal_ridge_alpha = 0
max_auc = 0
for alpha, auc in ridge_mean_auc.items():
    if auc > max_auc:
        max_auc = auc
        optimal_ridge_alpha = alpha
        
alphas = list(ridge_mean_auc.keys())

aucs = list(ridge_mean_auc.values())

plt.figure(figsize=(10, 6))  # Set the figure size

# Create the line plot
plt.plot(alphas, aucs, marker='o', linestyle='-', markersize=8, label='AUC')

# Add labels to the points
for i, txt in enumerate(aucs):
    plt.annotate(round(txt, 4), (alphas[i], aucs[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Set x-axis to be log scale
plt.xscale("log")

# Set labels and title
plt.xlabel("Alpha (log scale)")
plt.ylabel("AUC")
plt.title("AUC vs. Alpha")

# Show gridlines
plt.grid(True)

# Show legend
plt.legend()

# Display the plot
plt.show()
```

### A-9: Ridge Area Under Curve

```python
def ridge_area_under_curve(model_R, x_test: np.ndarray, y_test: np.ndarray) -> float:
    '''
        return area under the curve measurements of trained Ridge model used to find coefficients, i.e., model trained with optimal_aplha
        Also plot the ROC Curve using false_positive_rate, true_positive_rate as x and y axes calculated from roc_curve
    '''
    # Get the prediction scores
    y_pred = model_R.predict(x_test)
    
    # Assuming the positive class label is 1
    fpr, tpr, _ = roc_curve(y_test, y_pred, pos_label=1)  
    
     # Compute AUC score
    auc_score = roc_auc_score(y_test, y_pred)
    
    # Plot ROC curve
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc_score)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc="lower right")
    plt.show()
    
    return auc_score

ridge_auc = ridge_area_under_curve(model_R, X_test, y_test)
print(ridge_auc)
```

Below is a suggested sequence of steps you may want to think along for building regression and classification trees.

1. **Defining a criteria for splitting.**
    1. This criteria assigns a score to a split.
    1. For regression trees, this would be the mean squared error.
    2. For decision trees, this would be the Gini index or entropy.
2. **Create the split.**
    1. Split the dataset by iterating over all the rows and feature columns.
    2. Evaluate all the splits using the splitting criteria.
    3. Choose the best split.
3. **Build the tree.**
    1. Terminal nodes: decide when to stop growing a tree. This would be the maximum allowed depth of the tree or when a leaf is empty or has only 1 element.
    2. Recursive splitting: once a split is created, you can split it further recursively by calling the same splitting function on it.
    3. Building a tree: create a root node and apply recursive splitting on it.
4. **Make predictions with the tree.**
    1. For a given data point, make a prediction using the tree.

**Build the regression tree in the `TreeRegressor` class**.  

```python
#import libraries here if required
import pandas as pd 
import numpy as np 
```

```python
class TreeRegressor():
    def __init__(
        self, 
        data, 
        node_rule = "", 
        node_type = 'root', 
        min_samples_split_size = 2, 
        depth = 0, 
        max_depth=1
        ):
        
        self.data = data # last element of each row in data is the target variable
        self.X  = self.get_X() # Contains values of the predictors
        self.y  = self.get_Y() # Contains values of the target variable
        
        
        self.features = self.X.columns
        
        #Calculating the mean of target variable
        self.y_mean = np.mean(self.y)
        
        
        self.mse = self.mean_squared_error(self.y, self.y_mean)
        
        
        #Storing the number of observations in the node
        self.n_observations = len(self.y)

        # Saving the hyper parameters
        self.min_samples_split_size = min_samples_split_size
        self.max_depth = max_depth 
        
        #Default depth of current n|ode
        self.depth = depth 
        
        # Type of node
        self.node_type = node_type
        
        # Rule for spliting 
        self.node_rule = node_rule 
        
        # Setting the left node and right node to be pointing to None
        self.left_node = None 
        self.right_node = None 

        # Default values for splits
        self.best_feature_selected = None 
        self.best_selected_value = None 
    
    @staticmethod
    def moving_average(x: np.array, window: int) -> np.array:
        """
        Calculates the moving average of the given list. 
        """
        return np.convolve(x, np.ones(window), 'valid') / window

    def get_X(self):
        """
         Returns a list containing the values of target variable
        """
        features = self.data.iloc[:, :-1].columns
        return self.data[features]
        
        
    
    def get_Y(self):
        """
         Returns a list containing the values of target variable
        """
        target_values = self.data['Y'].values
        return target_values

        
    def find_best_split(self):
        """
            Given a set of features X and Target Variables Y find the best split for the tree.
        """
        
        #get the dataset 
        X_df = self.X.copy()
        X_df['Y'] = self.y
        
        #get mse value for base input dataset
        mse_base = self.mse
        
        selected_feature = None
        selected_split_value = None
        
        for feature in self.features:
            X_df_sorted = X_df.sort_values(feature)
            
            # calculate moving average of the sorted feature values
            X_avgs = self.moving_average(X_df_sorted[feature].unique(),2)

            for value in X_avgs:
                left_y = X_df_sorted[X_df_sorted[feature] < value]['Y'].values
                right_y = X_df_sorted[X_df_sorted[feature] >= value]['Y'].values
                
                # Getting the means 
                left_mean = np.mean(left_y)
                right_mean = np.mean(right_y)
                
                left_residual= left_y - left_mean 
                right_residual = right_y - right_mean
                
                combined_residual = np.concatenate((left_residual, right_residual), axis=None)

                # Calculating the mse 
                n = len(combined_residual)
                residual_squared = combined_residual ** 2
                r_sum = np.sum(residual_squared)
                mse_of_split = r_sum / n
                
                if mse_of_split < mse_base:
                    selected_feature = feature
                    selected_split_value = value
                    
                    mse_base = mse_of_split
        
        return selected_feature, selected_split_value 
                    
    # Build the recursive
    def build_tree(self):
        """
            Recursively build the Regression Tree 
        """
        data_df = self.X.copy()
        data_df['Y'] = self.y
        
        if (self.depth < self.max_depth) and (self.n_observations >= self.min_samples_split_size):
            selected_feature,selected_split_value = self.find_best_split()
            if selected_feature is not None:
                
                self.best_feature_selected = selected_feature
                self.best_selected_value = selected_split_value
                left_data_df, right_data_df = data_df[data_df[selected_feature] <= selected_split_value].copy(), data_df[data_df[selected_feature]> selected_split_value].copy()
                #creating the left and right node
                left_node = TreeRegressor(
                                left_data_df,
                                depth=self.depth + 1,
                                max_depth=self.max_depth,
                                min_samples_split_size = self.min_samples_split_size,
                                node_type='left_node',
                                node_rule=f"{selected_feature} <= {round(selected_split_value, 3)}"
                            )
                self.left_node = left_node
                self.left_node.build_tree()
                
                #creating the left and right node
                right_node = TreeRegressor(
                                right_data_df,
                                depth=self.depth + 1,
                                max_depth=self.max_depth,
                                min_samples_split_size = self.min_samples_split_size,
                                node_type='right_node',
                                node_rule=f"{selected_feature} <= {round(selected_split_value, 3)}"
                            )
                self.right_node = right_node
                
                self.right_node.build_tree()
                
                
    def print_info(self, width=4):
        """
        Method to print the infromation about the tree
        """
        # Defining the number of spaces 
        const = int(self.depth * width ** 1.5)
        spaces = "-" * const
        
        if self.node_type == 'root':
            print("Root")
        else:
            print(f"|{spaces}|{self.node_type}| Split rule: {self.node_rule}")
        print(f"{' ' * const}   | Node MSE: {round(self.mse, 2)}")
        print(f"{' ' * const}   | Node Observations Count: {self.n_observations}")
        print(f"{' ' * const}   | Node Prediction: {round(self.y_mean, 3)}")   

    def print_tree(self):
        """
        Prints the whole tree from the current node to the bottom
        """
        self.print_info() 
        
        if self.left_node is not None: 
            self.left_node.print_tree()
        
        if self.right_node is not None:
            self.right_node.print_tree()
          

    @staticmethod
    def mean_squared_error(y_actual, y_pred):
        """
            Calculates and returns the mean squared error
        """
        obs_size = len(y_actual)
       
        residuals = y_actual - y_pred
        residual_squares = residuals**2
        r_sum = np.sum(residual_squares)
        return r_sum / obs_size
    
    def batch_predict(self, input_data: pd.DataFrame):
        """
        Batch prediction method
        """
        batch_predictions = []

        for _, row in input_data.iterrows():
            input_values = {}
            for feature in self.features:
                input_values[feature] = row[feature]

            prediction = self.predict_value(input_values)
            batch_predictions.append(prediction)

        return batch_predictions
    
    def predict_value(self, input_values: dict) -> int:
        """
        Method to predict a value for a given set of features
        """
        current_node = self

        while current_node.right_node or current_node.left_node:
            best_feature = 'X'  # Replace with your actual best feature selection logic
            best_value = current_node.best_selected_value

            if current_node.n_observations < current_node.min_samples_split_size:
                break 

            if input_values.get(best_feature) < best_value:
                if current_node.left_node is not None:
                    current_node = current_node.left_node
            else:
                if current_node.right_node is not None:
                    current_node = current_node.right_node

        return current_node.y_mean
    
tree_regressor_model = TreeRegressor(data_regress_df,max_depth=4,min_samples_split_size=1)
tree_regressor_model.build_tree()
tree_regressor_model.print_tree()
X = list(np.arange(0,1.01,0.01))
X = pd.DataFrame(X, columns=['X'])
y = tree_regressor_model.batch_predict(X)
```

```python
#import required libraries
from collections import Counter
from sklearn.metrics import accuracy_score
```

```python
# TreeClassifier is a derived class of TreeRegressor
# Don't forget to create a method to make predictions using your tree.

class TreeClassifier(TreeRegressor):
    def __init__(self, 
        data, 
        node_rule = "", 
        node_type = 'root', 
        min_samples_split_size = 2, 
        depth = 0, 
        max_depth=1):
        super().__init__(
                data,
                node_rule = node_rule, 
                node_type = node_type, 
                min_samples_split_size = min_samples_split_size,
                depth = depth, 
                max_depth=max_depth,)
        
        #find the number of observations belonging to the two classes
        self.class_counts = Counter(self.y)
        
        #save the gini index calcualted for the node
        self.gini_index = self.gini_index()
        
        # Sorting the class counts so that we store the most common occuring class
        sorted_counts = list(sorted(self.class_counts.items(), key=lambda item: item[1]))

        # getting the last item which is the class having highest observations
        y_pred = None
        if len(sorted_counts) > 0:
            y_pred = sorted_counts[-1][0]

        # The class having highest observations will be the most frequent occuring class and hence that will be prediction for the node.
        self.y_pred = y_pred 
        
    def find_best_split(self) -> tuple:
        """
        Given the X features and Y targets finds the best split 
        for a classifier decision tree
        """
        #get the dataset 
        X_df = self.X.copy()
        X_df['Y'] = self.y
        
        # Getting the GINI impurity for the base input 
        GINI_base_impurity = self.gini_index

        # Finding which split yields the best GINI gain 
        max_gini_gain = 0
        
        selected_feature = None
        selected_split_value = None
        
        for feature in self.features:
            X_df_sorted = X_df.sort_values(feature)
            
            # calculate moving average of the sorted feature values
            X_avgs = self.moving_average(X_df_sorted[feature].unique(),2)

            for value in X_avgs:
                # Spliting the dataset 
                left_counts,right_counts  = Counter(X_df_sorted[X_df_sorted[feature]<value]['Y']),Counter(X_df_sorted[X_df_sorted[feature]>=value]['Y'])
                
                # Getting the Y distribution from the dicts
                y0_left, y1_left, y0_right, y1_right = left_counts.get(-1, 0), left_counts.get(1, 0), right_counts.get(-1, 0), right_counts.get(1, 0)
              
                # Getting the left and right gini impurities
                gini_left = self.GINI_impurity(y0_left, y1_left)
                gini_right = self.GINI_impurity(y0_right, y1_right)

                # Getting the obs count from the left and the right data splits
                n_left = y0_left + y1_left
                n_right = y0_right + y1_right

                # Calculating the weights for each of the nodes
                w_left = n_left / (n_left + n_right)
                w_right = n_right / (n_left + n_right)

                # Calculating the weighted GINI impurity
                wGINI = w_left * gini_left + w_right * gini_right

                # Calculating the GINI gain 
                GINI_gain = GINI_base_impurity - wGINI

                # Checking if this is the best split so far 
                if GINI_gain > max_gini_gain:
                    selected_feature = feature
                    selected_split_value = value 

                    # Setting the best gain to the current one 
                    max_gini_gain = GINI_gain

        return (selected_feature, selected_split_value)

    
       # Build the recursive
    def build_tree(self):
        """
            Recursively build the Regression Tree 
        """
        data_df = self.X.copy()
        data_df['Y'] = self.y
        
        if (self.depth < self.max_depth) and (self.n_observations >= self.min_samples_split_size):
            
            selected_feature,selected_split_value = self.find_best_split()
            
            if selected_feature is not None:
                
                self.best_feature_selected = selected_feature
                self.best_selected_value = selected_split_value
                
                left_data_df, right_data_df = data_df[data_df[selected_feature] <= selected_split_value].copy(), data_df[data_df[selected_feature]> selected_split_value].copy()
                
                #creating the left and right node
                left_node = TreeClassifier(
                                left_data_df,
                                depth=self.depth + 1,
                                max_depth=self.max_depth,
                                min_samples_split_size = self.min_samples_split_size,
                                node_type='left_node',
                                node_rule=f"{selected_feature} <= {round(selected_split_value, 3)}"
                            )
                self.left_node = left_node
                self.left_node.build_tree()
                
                #creating the left and right node
                right_node = TreeClassifier(
                                right_data_df,
                                depth=self.depth + 1,
                                max_depth=self.max_depth,
                                min_samples_split_size = self.min_samples_split_size,
                                node_type='right_node',
                                node_rule=f"{selected_feature} <= {round(selected_split_value, 3)}"
                            )
                self.right_node = right_node
                
                self.right_node.build_tree()
                
    # Calculate the Gini index for a split dataset
    def gini_index(self):
        """
        Function to calculate the GINI impurity of a node 
        """
        # Getting the counts of each class
        class1_count, class2_count = self.class_counts.get(-1, 0), self.class_counts.get(1, 0)

        # Getting the GINI impurity
        return self.GINI_impurity(class1_count, class2_count)
    
    @staticmethod
    def GINI_impurity(class1_count: int, class2_count: int) -> float:
        """
        Given the observations of a binary class calculate the GINI impurity
        """
        # Ensuring the correct types
        if class1_count is None:
            class1_count = 0

        if class2_count is None:
            class2_count = 0

        # Getting the total observations
        n = class1_count + class2_count
        
        # return 0 as the GINI impurity if n is 0
        if n == 0:
            return 0.0

        # Getting the class proportions
        prop1 = class1_count / n
        prop2 = class2_count / n
        
        # Calculating GINI 
        gini_impurity = 1 - (prop1 ** 2 + prop2 ** 2)
        
        # Returning the gini impurity
        return gini_impurity
    
    def print_info(self, width=4):
        """
        Method to print the infromation about the tree
        """
        # Defining the number of spaces 
        const = int(self.depth * width ** 1.5)
        spaces = "-" * const
        
        if self.node_type == 'root':
            print("Root")
        else:
            print(f"|{spaces}|{self.node_type}| Split rule: {self.node_rule}")
        
        print(f"{' ' * const}   | GINI impurity of the node: {round(self.gini_index, 2)}")
        print(f"{' ' * const}   | Class distribution in the node: {dict(self.class_counts)}")
        print(f"{' ' * const}   | Predicted class: {self.y_pred}")   
            
    def batch_predict(self, input_data: pd.DataFrame):
        """
        Batch prediction method
        """
        batch_predictions = []

        for _, row in input_data.iterrows():
            input_values = {}
            for feature in self.features:
                input_values[feature] = row[feature]

            prediction = self.predict_value(input_values)
            batch_predictions.append(prediction)

        return batch_predictions
    
    def predict_value(self, values: dict) -> int:
        """
        Method to predict the values for given a set of features
        """
        curr = self
        while curr.right_node or curr.left_node:
            # Traversing the nodes all the way to the bottom
            best_feature = curr.best_feature_selected
            best_value = curr.best_selected_value

            if curr.n_observations < curr.min_samples_split_size:
                break 

            if (values.get(best_feature) < best_value):
                if self.left_node is not None:
                    curr = curr.left_node
            else:
                if self.right_node is not None:
                    curr = curr.right_node
            
        return curr.y_pred

classifier_model = TreeClassifier(data_classification_df,max_depth=8,min_samples_split_size=4)
classifier_model.build_tree()
classifier_model.print_tree()
y_pred = classifier_model.batch_predict(data_classification_df[columns])
```

```python
# Plot the data
plt.figure()
plt.scatter(data_class[:, 0], data_class[:, 1], c=y_pred, cmap='bwr')
plt.xlabel("Features, x1")
plt.ylabel("Features, x2")
plt.show()
```

**Plot the classification accuracy at each depth of your tree.**

```python
accuracy_at_depth = []

for max_depth in range(1, 8):  # Depth values from 1 to 7
    # Create a new classifier model with the specified max_depth
    classifier_model = TreeClassifier(data_classification_df, max_depth=max_depth, min_samples_split_size=3)
    classifier_model.build_tree()
    
    # Make predictions using the current model
    y_pred = classifier_model.batch_predict(data_classification_df[columns])
    
    # Calculate accuracy and store it
    accuracy = accuracy_score(data_classification_df['Y'], y_pred)
    accuracy_at_depth.append((max_depth, accuracy))

# Extract depth and accuracy values for plotting
depths, accuracies = zip(*accuracy_at_depth)

# Plot depth vs. accuracy
plt.figure()
plt.plot(depths, accuracies, marker='o', linestyle='-')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.title('Depth vs. Accuracy')
plt.grid(True)
plt.show()
```

```python
## # Drop those columns which has null values
df.dropna(inplace=True)

# Drop unwanted columns in feature space
features = df.drop(columns=['NewLeague','Player']) #X
features.head()
#One hot encoding of non-numerical features
non_numerical_features = features.select_dtypes(exclude=['int64','float64'])
numerical_features = features.select_dtypes(include=['int64','float64'])
encoded_features = pd.get_dummies(non_numerical_features, columns=non_numerical_features.columns)
processed_features = pd.concat([encoded_features, numerical_features],axis=1)

#Standard Scaling since Ridge Regression is sensitive to the scale of data.
scaler = StandardScaler()
scaled_features = scaler.fit_transform(processed_features)
scaled_features_df = pd.DataFrame(scaled_features, columns=processed_features.columns)


#Replace the non-numerical classes into numerical values
labels = df['NewLeague'] #y
labels.replace('N',1,inplace=True)
labels.replace('A',0,inplace=True)
labels_df = pd.DataFrame(labels, columns=['NewLeague'])

#New Data Frame with scaled values and coded target values.
scaled_features_df.reset_index(drop=True, inplace=True)
labels_df['NewLeague'].reset_index(drop=True, inplace=True)
scaled_features_df.head()
```

```python
#Splitting the data set into train, Validation and Test sets
X_train, X_test, y_train, y_test = train_test_split(scaled_features_df, labels, test_size = 0.15, random_state = 42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.176, random_state=1) 
print("Training Data Size",X_train.shape)
print("Validation Data Size", X_valid.shape)
print("Test Data Size",X_test.shape)
```

```python
import xgboost as xgb
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error

def train_XGBoost( x_train: np.ndarray, y_train:np.ndarray, x_val: np.ndarray, y_val: np.ndarray)-> dict:
    '''
        Instantiate an object of XGBoost, train the model object using training data for the given N-bootstraps iterations.
        In each iteration train the model for all alpha_vals, compute the auc values with the validation set and append the values to aucs_xgboost.
        Return the "xgboost_mean_auc" dictionary which stores the mean auc for each alpha.
    '''
    n_bootstraps = int(1e2)
    alpha_vals = [1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3]
    aucs_xgboost = {alpha:[] for alpha in alpha_vals}
    best_mean_auc = 0.0
    for _ in range(n_bootstraps):
        x_train_sample, y_train_sample = resample(x_train, y_train)
        for alpha in alpha_vals:
            # Create and train an XGBoost model with L2 regularization for each bootstrap iteration
            model = xgb.XGBClassifier(reg_alpha=0,reg_lambda=alpha)
            model.fit(x_train_sample, y_train_sample)

            # Predict on the validation set
            y_pred = model.predict(x_val)

            # Calculate the AUC score
            xgboost_auc = roc_auc_score(y_val, y_pred)

            aucs_xgboost[alpha].append(xgboost_auc)

        xgboost_mean_auc = {alpha: np.mean(aucs) for alpha, aucs in aucs_xgboost.items()}

    return xgboost_mean_auc

# Find the best alpha using train_XGBoost function
xgboost_mean_auc = train_XGBoost(X_train, y_train, X_valid, y_valid)

max_auc = 0
best_alpha = 0

for alpha, auc in xgboost_mean_auc.items():
    if auc > max_auc:
        max_auc = auc
        best_alpha = alpha
        
alphas = list(xgboost_mean_auc.keys())

aucs = list(xgboost_mean_auc.values())

plt.figure(figsize=(10, 6))  # Set the figure size

# Create the line plot
plt.plot(alphas, aucs, marker='o', linestyle='-', markersize=8, label='AUC')

# Add labels to the points
for i, txt in enumerate(aucs):
    plt.annotate(round(txt, 4), (alphas[i], aucs[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Set x-axis to be log scale
plt.xscale("log")

# Set labels and title
plt.xlabel("Alpha (log scale)")
plt.ylabel("AUC")
plt.title("AUC vs. Alpha")

# Show gridlines
plt.grid(True)

# Show legend
plt.legend()

# Display the plot
plt.show()

# Print the best alpha
print(f"Best Alpha: {best_alpha}")
```
